# Results figures (Phase 4/5)

Loads `results/frontier_points.json` (produced by `scripts/04_eval_frontier.py`) and
renders the headline sparsity↔fidelity frontier plot, plus the training curves if you
saved a history. Use these figures in the README/blog writeup.

```python
import json, pathlib
import matplotlib.pyplot as plt

pts = json.load(open('../results/frontier_points.json'))
pts = sorted(pts, key=lambda p: p['l0'])
pts[:2]
```

```python
fig, ax = plt.subplots(figsize=(6,4.5))
ax.plot([p['l0'] for p in pts], [p['variance_explained'] for p in pts], 'o-')
for p in pts:
    ax.annotate(f"λ={p['l1_coeff']:.0e}", (p['l0'], p['variance_explained']),
                textcoords='offset points', xytext=(5,5), fontsize=8)
ax.set_xlabel('L0 (active features / token)')
ax.set_ylabel('variance explained')
ax.set_title('Sparsity ↔ fidelity frontier')
ax.grid(alpha=0.3); fig.tight_layout()
fig.savefig('../results/figures/frontier.png', dpi=150)
fig
```

```python
# If CE-loss-recovered was computed (--ce), plot it too.
if all('ce_loss_recovered' in p for p in pts):
    fig, ax = plt.subplots(figsize=(6,4.5))
    ax.plot([p['l0'] for p in pts], [p['ce_loss_recovered'] for p in pts], 'o-', color='green')
    ax.set_xlabel('L0'); ax.set_ylabel('CE-loss recovered'); ax.grid(alpha=0.3)
    ax.set_title('Sparsity ↔ behaviour-preservation frontier'); fig.tight_layout()
    fig.savefig('../results/figures/frontier_ce.png', dpi=150)
    display(fig)
else:
    print('No CE-loss-recovered in points; re-run scripts/04_eval_frontier.py --ce')
```
